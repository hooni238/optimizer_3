```python
def dis(a,b,c,x,y,z):
    return (int(a-x)**2+int(b-y)**2+int(c-z)**2)**0.5
```

```python
m1 = [[], [], []]
m1Ind = []
x, y, z = 0, 0, 0
indis = 0
for i in range(len(X)):
    if X[i] > x:
        x, y, z = X[i], Y[i], Z[i]
        indis = i
        
m1[0].append(x)
m1[1].append(y)
m1[2].append(z)
m1Ind.append(indis)

thresh = 110

XX, YY, ZZ = [], [], []

flag = 1
while flag:
    flag -= 1
    for i in range(len(X)):
        for j in range(1):
            #print('\n\n', X[i],Y[i],Z[i], m1[0][j], m1[1][j], m1[2][j], dis(X[i],Y[i],Z[i], m1[0][j], m1[1][j], m1[2][j]))
            if dis(X[i],Y[i],Z[i], m1[0][j], m1[1][j], m1[2][j]) < thresh:
                m1[0].append(X[i])
                m1[1].append(Y[i])
                m1[2].append(Z[i])
                m1Ind.append(i)
            else:
                XX.append(X[i])
                YY.append(Y[i])
                ZZ.append(Z[i])

#X, Y, Z = XX, YY, ZZ
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')
#ax.view_init(azim=90)
ax.scatter(m1[0], m1[1], m1[2])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
t = 6
ax.scatter(m1[0][t], m1[1][t], m1[2][t], marker='*', s = 300)
plt.show()
```
